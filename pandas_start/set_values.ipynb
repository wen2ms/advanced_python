```python
import pandas as pd
import numpy as np

dates = pd.date_range('20250101', periods=6)

dataframe = pd.DataFrame(np.arange(24).reshape((6, 4)), index=dates, columns=['A', 'B', 'C', 'D'])

print_newline = lambda data: print(data, '\n')

print_newline(dataframe)

dataframe.iloc[2, 2] = -1

dataframe.loc['20250101', 'A'] = -2

print_newline(dataframe)

print_newline(dataframe[dataframe.A > 0])

print_newline(dataframe.A[dataframe.A > 0])

print_newline(dataframe.B[dataframe.A > 0])

dataframe['E'] = np.nan
print_newline(dataframe)

dataframe['F'] = pd.Series(np.arange(1, 7), index=dates)
dataframe
```
